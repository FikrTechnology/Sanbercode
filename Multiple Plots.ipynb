```python
import matplotlib.pyplot as plt
import numpy as np
```

<h2>No 1 : Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output :

```python
x = np.linspace(2*-np.pi, 2*np.pi, 200)
tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
fig, ax = plt.subplots(ncols = 3, figsize = (12,8), sharex = True, sharey = True)

ax[0].plot(x, tan, label = 'TanX')
ax[1].plot(x, sin, color = 'red', label = 'SinX')
ax[2].plot(x, cos, color = 'green', label = 'CosX')

ax[0].set_title('Grafik TanX')
ax[1].set_title('Grafik SinX')
ax[2].set_title('Grafik CosX')

fig.legend()
plt.show()
```

<h2>No 2 : Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output :

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
fig, ax = plt.subplots(figsize=(16,8))

ax.plot(x, y)

ins1 = ax.inset_axes([-0.2, -0.2, 0.4, 0.4])
ins1.plot(x, y)
ins1.plot(x, y2, color='r')

ins2 = ax.inset_axes([0.9, -0.2, 0.4, 0.4])
ins2.plot(x, y)
ins2.plot(x, y3, color='g')

ins3 = ax.inset_axes([0.9, 0.9, 0.4, 0.4])
ins3.plot(x, y)
ins3.plot(x, y4, color='black')

ins4 = ax.inset_axes([-0.2, 0.9, 0.4, 0.4])
ins4.plot(x, y)
ins4.plot(x, y3, color='y')

ax.set_title('Grafik CosX')
ins1.set_title('Grafik CosX^2')
ins2.set_title('Grafik CosX^3')
ins3.set_title('Grafik CosX^4')
ins4.set_title('Grafik CosX^5')
plt.show()
```

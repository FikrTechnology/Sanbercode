<h1>Quiz 2 : Pemahaman Machine Learning</h1>

Pelajarilah secara garis besar suatu model/algoritma machine learning, kemudian applikasikan untuk membuat model bagi data (variable x, y) di bawah ini. kemudian buatlah prediksi terhadap data training dan data baru di interval 20-30.

- Plot data asli, data hasil prediksi terhadap data training, data hasil prediksi terhadap data baru
- Tunjukan beberapa parameter yang dimiliki model yang telah kalian buat

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression # 1
from sklearn.tree import DecisionTreeRegressor
from sklearn.tree import export_graphviz

rng = np.random.RandomState(42)
x = 20 * rng.rand(50)
y = x**2 + 2 * x + - 1 + rng.randn(50)

# data yang akan di prediksi
x_new = np.arange(20, 30, 0.5)

fig, ax = plt.subplots(figsize=(12, 6))
ax.scatter(x, y)
ax.set_xlabel('X-Value')
ax.set_ylabel('Y-Value')
plt.show()
```

Contoh Jawaban Hasil Pembuatan Model dengan Algoritma DecisionTreeRegressor (Jawaban tidak haru sama) :



```python
# Menggunakan Teknik Linear Regression
# 2
model_lr = LinearRegression(fit_intercept=True)

# 3
x_matriks = x[:, np.newaxis]

# 4 Data telah di training
model_lr.fit(x_matriks, y)

#dalam test data perlu kita memerlukan data baru
x_test = np.linspace(20, 30, 15)
x_test = x_test[:, np.newaxis]

# 5 menggunakan method .predict()
y_test = model_lr.predict(x_test)

y_train = model_lr.predict(x_matriks)
plt.scatter(x, y)
plt.plot(x, y_train, color='y', label='Prediksi terhadap data Training')
plt.plot(x_test, y_test, color='g', label='Prediksi terhadap data Baru')
plt.legend()
plt.show()
```

```python
# Teknik Menggunakan Decision Tree Model
# 2
regressor = DecisionTreeRegressor(max_depth = 2)

x_matriks = x[:, np.newaxis]

regressor.fit(x_matriks, y)

x_test = np.arange(20, 30, 0.5)
x_test = x_test[:, np.newaxis]

y_test = regressor.predict(x_test)
y_train = regressor.predict(x_matriks)

fig, ax = plt.subplots(figsize=(16, 8))
ax.scatter(x, y, label='Data Asli')
ax.scatter(x, y_train, color='y', label='Prediksi terhadap data Training')
ax.plot(x_test, y_test, color='g', label='Prediksi terhadap data Baru')

plt.legend()
plt.show()
```
